# CLEANING DATASET

## Cek Missing Value

```python
#Cek missing value 
missing_counts = df.isnull().sum()
print(missing_counts[missing_counts>0])
```

```python
#drop the missing value 
df.dropna(inplace = True)
df
```

```python
df.info()
```

## Change Format

```python
#Mengubah Format float menjadi integer
df[['id','price', 'host_id', 'minimum_nights', 'available_365']] = df[['id','price', 'host_id', 'minimum_nights', 'available_365']].astype(int)

#Mengubah format tanggal menjadi datetime
df['date'] = pd.to_datetime(df['date'])
```

## Cek Outlier

```python
#cek outlier 
cek_out_price = px.box(df, x = 'price', width = 800, height = 500)
cek_out_price.show()
```

  Persebaran harga penyewaan Airbnb memiliki harga paling rendah sebesar 15 dollar dan paling mahal 12500 dollar. Rata-rata harga berkisar dari 47 - 18 dollar. Sedangkan untuk harga di atas 224 dollar, jumlahnya tidak terlalu banyak, dan hanya satu penginapan yang berharga paling mahal.

  Melihat persebaran harga penyewaan yang terlalu jauh, maka diperlukan cleaning outlier. 



```python
#Menghapus outlier 
Q1 = df['price'].quantile(0.25) 
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1 

#membuat bolean mask untuk nilai non-outlier 
price_cl_mask = (df['price'] >= Q1 - 1.5 * IQR) & (df['price'] <= Q3 + 1.5 * IQR)

#terapkan dataframe mask untuk menyimpan hanya nilai yang telah dicleaning dr outlier
df_cl = df[price_cl_mask]

```

```python
# plot boxplot 
dist_harga_clean = px.box(df_cl, x='price', width=1000)
# modify title
dist_harga_clean.update_layout(title="<b>Distribusi Harga Penginapan di Singapura</b>", 
                               title_font=dict(size=20, family='montserrat'))
dist_harga_clean.show()
```

## Data Valiadation (Cek Wilayah)

```python
#Tinjau peta yang menunjukkan persebaran Airbnb Singapura 
# plotting map using plotly
fig_map = px.scatter_mapbox(df_cl, lat = 'latitude', lon = 'longitude', color = 'neighbourhood_group', zoom = 10, mapbox_style='open-street-map', width = 750)

# setting title for the map
fig_map.update_layout(title = 'Countries', title_x = 0.5)

fig_map.show()
```

Peta menunjukkan persebaran Airbnb yang seharusnya berada di wilayah Singapura. Namun, terdapat beberapa listing yang berada di luar wilayah Singapura, yaitu Malaysia. Batas antara Singapura dan Malaysia ditandai oleh garis ungu putus-putus. 

Oleh karena itu, listing yang berada di sebelah utara Singapura akan dihapus dan menyisakan data valid yang berisi listing Airbnb Singapura saja. 

```python
#Menghapus outlier latitude dan longitude 
#definisikan batas latitude dan longitude 
lat_min, lat_max = 1.27316, 1.449275
lon_min, lon_max = 103.6934, 103.9671

#membuat masking untuk batas latitude dan lonmgitude Singapura
map_mask = (df_cl['latitude'] >= lat_min) & (df_cl['latitude'] <= lat_max) & \
       (df_cl['longitude'] >= lon_min) & (df_cl['longitude'] <= lon_max)

#drop rows diluar batas 
df_cl = df_cl[map_mask]

#reset kembali index di dataframe
df_cl = df_cl.reset_index(drop = True)

# plotting map using plotly
fig_map = px.scatter_mapbox(df_cl, lat = 'latitude', lon = 'longitude', color = 'neighbourhood_group', zoom = 10, mapbox_style='open-street-map', width = 750)

# setting title for the map
fig_map.update_layout(title = 'Countries', title_x = 0.5)

fig_map.show()
```

### Perbandingan Tipe Penginapan yang paling banyak diminati berdasarkan wilayah

```python
import plotly.graph_objects as go
import numpy as np

# Group the listings by neighborhood group and room type, and count the number of listings for each group
listing_room_rvw = df_cl.groupby(['neighbourhood_group', 'room_type']).size().reset_index(name='count')

# Calculate the total number of listings for each neighborhood group
total_listings = df_cl.groupby('neighbourhood_group').size().reset_index(name='total')

# Merge the total listings with the room type data
listing_room_rvw = pd.merge(listing_room_rvw, total_listings, on='neighbourhood_group')

# Calculate the percentage for each room type in each neighborhood group
listing_room_rvw['percent'] = listing_room_rvw['count'] / listing_room_rvw['total'] * 100

# Sort the data by count in descending order
listing_room_rvw = listing_room_rvw.sort_values(by='count', ascending=False)

# Create a color scale that goes from light to dark
color_scale = np.linspace(0, 255, len(listing_room_rvw['room_type'].unique())).astype(int)

# Create the figure
fig_tipe_reg = go.Figure()

# Iterate over the room types and add a bar chart for each one
for i, room_type in enumerate(listing_room_rvw['room_type'].unique()):
    # Get the count and percentage for the current room type and neighborhood group
    room_type_data = listing_room_rvw[listing_room_rvw['room_type'] == room_type]
    
    # Get the color for the current bar (darker for higher counts)
    count_index = np.where(room_type_data['count'] == room_type_data['count'].max())[0][0]
    color = f'rgb({200}, {color_scale[i]}, {color_scale[i]})' if i == count_index else f'rgb({255//2}, {color_scale[i]//2}, {color_scale[i]//2})'
    
    # Add the bar chart to the figure
    fig_tipe_reg.add_trace(go.Bar(x=room_type_data['neighbourhood_group'], 
                                  y=room_type_data['count'], 
                                  name=room_type, 
                                  marker_color=color,
                                  text=np.round(room_type_data['percent'], 2),
                                  textposition='auto'))
 
# Update the figure layout and show the plot
fig_tipe_reg.update_layout(xaxis_title='Neighborhood Group', yaxis_title='Count', barmode='group')
fig_tipe_reg.show()
```

Bar chart gambar 7 tersebut menunjukkan tipe penginapan mana yang paling laris dari kelima wilayah yang ada di Singapura. Berdasarkan data tersebut dapat disimpulkan bahwa Private room menjadi tipe kamar paling diminati di seluruh wilayah di Singapura. 

Merujuk pada data di atas, memilih tipe Private room menjadi rekomendasi terbaik jika ingin terjun dalam bisnis penyewaan Airbnb.
