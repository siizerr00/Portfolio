```python
#Import LIbraries
import pandas as pd
import pandasql as ps
import numpy as np 
from datetime import datetime as dt 

#Library EDA
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Import Dataset 
df = pd.read_csv('https://raw.githubusercontent.com/siizerr00/Portfolio/Data-Analyst/SQL%20Project/Jubelio/q3_q4_review_raw.csv', encoding='unicode_escape')
df
```

```python
df = df.fillna(0)
df['price_deduction'] = df['price_deduction'].astype(int)
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
df
```

```python
df
```

```python
import calendar

# membuat kolom baru 'purchase_month' berdasarkan kolom 'purchase_date'
df['purchase_month'] = df['purchase_date'].dt.month

# menghitung total promo per bulan
total_promo = df.groupby('purchase_month')['price_after_promo'].sum().reset_index()

# urutan bulan
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# ubah format bulan
total_promo['purchase_month'] = total_promo['purchase_month'].apply(lambda x: calendar.month_name[x])
total_promo = total_promo.set_index('purchase_month').reindex(month_order).reset_index()

#plot
fig_promo = px.line(total_promo, x='purchase_month', y='price_after_promo', labels={'purchase_month':'Bulan', 'price_after_promo':'Total Promo'}, title='Total Promo per Bulan', markers=True, color_discrete_sequence=["#943455"])
fig_promo.update_layout(title='Total Promo per Bulan')
fig_promo.show()
```

```python
# tambahkan kolom 'purchase_month' yang berisi nama bulan dari tanggal pembelian
df['purchase_month'] = pd.to_datetime(df['purchase_date']).dt.month_name()

# tambahkan kolom 'with_promo' yang bernilai 1 jika ada promo dan 0 jika tidak ada promo
df['with_promo'] = (df['promo_id'] != 0).astype(int)

# tambahkan kolom 'with_promo' yang bernilai 'tidak promo' jika tidak ada promo dan 'promo' jika ada promo
df['with_promo'] = np.where(df['promo_id'] != 0, 'promo', 'tidak promo')

# menghitung total pembelian per bulan
total_sales = df.groupby(['purchase_month'])['total_price'].sum().reset_index()

# menghitung rasio pembelian dengan dan tanpa promo per bulan
monthly_sales = df.groupby(['purchase_month', 'with_promo']).size().reset_index(name='count')
promo_sales = monthly_sales.pivot(index='purchase_month', columns='with_promo', values='count').reset_index()
promo_sales['total'] = promo_sales['promo'] + promo_sales['tidak promo']
promo_sales['promo'] = promo_sales['promo'] / promo_sales['total']
promo_sales['tidak promo'] = promo_sales['tidak promo'] / promo_sales['total']

# urutkan data berdasarkan urutan bulan
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
promo_sales['purchase_month'] = pd.Categorical(promo_sales['purchase_month'], categories=month_order, ordered=True)
promo_sales = promo_sales.sort_values('purchase_month')

# membuat categorical bar chart
fig = px.bar(promo_sales, x='purchase_month', y=['tidak promo', 'promo'], color_discrete_sequence=['#943455', '#2E86C1'])
fig.update_layout(title='Rasio Pembelian dengan dan Tanpa Promo per Bulan', yaxis_tickformat='%')
fig.update_xaxes(title='Bulan', type='category', categoryorder='array', categoryarray=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
fig.update_yaxes(title='Persentase (%)')
fig.update_layout(yaxis_tickformat='.0%')




fig.show()
```
